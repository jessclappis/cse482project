```python
from surprise import Dataset
from surprise import Reader
# data = Dataset.load_builtin(name=u'ml-1m')
reader = Reader(line_format='user item rating', sep='\t')
data = Dataset.load_from_file("ratings.tsv", reader=reader)
data.raw_ratings
data.raw_ratings+['1','1','3.0'] #append whatever the input is
```

```python
import pandas as pd
movies = pd.read_csv("movies.csv", header=None)
movies
```

```python
from surprise import NMF 
algo = NMF (n_factors=4, n_epochs=100, random_state=1)
trainSet = data.build_full_trainset()
algo.fit(trainSet)
```

```python
predictions = []
#have i[0] and i[1] be the current user and movie id
for index, row in movies.iterrows():
    pred = algo.predict('19074',row[1], r_ui=4, verbose=True)
    predictions.append(pred)
```

```python
sortpred = sorted(predictions, key=lambda pred : pred[3])
sortpred = sortpred[-10:]
sortpred
```
